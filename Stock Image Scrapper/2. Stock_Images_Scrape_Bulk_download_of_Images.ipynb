### **3. Defining Dynamic Naming for Images**

```python
ref = 100000
for i in range(1,11):
    print('A' + str(i + ref)[1:])
```

### **4. Defining Image Downloader function**

```python
def download(link, img_path ):

    res = requests.get(link)         # Sending the Request
    fd = open(img_path,'wb')         # Getting Image in WB Format
    fd.write(res.content)            # Saving the Image
    fd.close()
```

```python
# Creating the Imgs folder in colab
folder_path = '/content/Imgs'
os.makedirs(folder_path)
```

### **5. Downloading all Images**

```python
paths   = []
img_ids = []

ref = 100000
i   = 1

for link in tqdm(df['img_link']):

    img_id = 'A' + str(i + ref)[1:]          # Creating Image ID
    i += 1

    img_path = '/content/Imgs/' + img_id + '.jpg'    # Creating Image Path

    download(link, img_path)                # Downloading the Image

    paths.append(img_path)
    img_ids.append(img_id)
```

```python
print(len(paths))
print(len(img_ids))
```

### **6. Saving Ids and Paths in DataFrame**

```python
df['img_id']   = img_ids
df['img_path'] = paths

df.head()
```

### **7. Saving Final Dataset**

```python
df.to_csv('imgs.csv', index = False)
```
